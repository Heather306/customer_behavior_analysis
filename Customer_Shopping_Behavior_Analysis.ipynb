```python
import pandas as pd

df = pd.read_csv("customer_shopping_behavior_dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
print(df.columns.tolist())
```

```python
labels = ['Young Adult', 'Adult', 'Middle-age', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Daily': 1,
    'Weekly': 7,
    'Monthly': 30
}

df['purchase_frequency_days'] = df['visit_frequency'].map(frequency_mapping)
```

```python
df[['visit_frequency', 'purchase_frequency_days']].head(10)
```

```python
df[['online_shopping','discount_used']].head(10)
```
